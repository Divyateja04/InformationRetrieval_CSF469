```python
import indexer
```

```python
index = indexer.load_index_in_memory("../../nfcorpus/raw/")[0]
index
```

```python
def retrieve_docs(entities, index):
    retrieved_docs = dict()
    for entity in entities:
        # Retrieve the documents for the entity from the index
        docs = index.get(entity, [])
        # Store the retrieved documents in the dictionary
        retrieved_docs[entity] = docs
    return retrieved_docs

# Define the list of entities you are interested in
entities = ["aging"]

# Call the function with your entities and index
retrieved_docs = retrieve_docs(entities, index)
print(retrieved_docs)
# Now, `retrieved_docs` is a dictionary where the keys are the entities and the values are the lists of documents where the entities occur
```

```python
def rank_docs(query, index):
    # Split the query into entities
    query_entities = query.split()

    # Initialize a dictionary to store the scores for each document
    doc_scores = {}

    # Iterate over the entities in the query
    for entity in query_entities:
        # Retrieve the documents for the entity from the index
        docs = index.get(entity, [])

        # Iterate over the documents
        for doc in docs:
            # If the document is not in the scores dictionary, add it with a score of 0
            if doc['doc_id'] not in doc_scores:
                doc_scores[doc['doc_id']] = 0

            # Add the term frequency to the document's score
            doc_scores[doc['doc_id']] += doc['term_freq']

    # Sort the scores in descending order and return them
    sorted_scores = sorted(doc_scores.items(), key=lambda item: item[1], reverse=True)
    return sorted_scores

# Define a query
query = "why deep fried foods may cause cancer"

# Rank the documents based on the query
sorted_scores = rank_docs(query, index)

# Print the scores
for doc_id, score in sorted_scores:
    print(f"Document ID: {doc_id}, Score: {score}")
```
